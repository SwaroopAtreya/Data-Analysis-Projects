```python
def remove_outliers(col_name):
    sorted(col_name)
    Q1, Q3 = col_name.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper  = Q3 + 1.5*IQR
    return lower, upper
```

```python
low,high = remove_outliers(df['K'])
```

```python
low
```

```python
high
```

```python
df['K'] = np.where(df['K']>high,high,df['K'])
```

```python
sns.boxplot(y='K',data =df)
# these all are the preprocessing operations and the data is highly qualitative and ready for data analysis we saw the null report, duplicate values and removed outliers
```
